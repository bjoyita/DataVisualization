**Importing Libraries**


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

**Loading data**
*Room temperature Thermoelectric proproperties of 204 compositions*

```python
ds = pd.read_csv('TE_props_correlation1.csv')
ds
```

```python
df=pd.DataFrame(ds)
```

**Converting to dataframe**

```python
df
```

**PANDAS: Correlation Matrix**

```python
matrix= df.corr()
matrix
```

With the background_gradient function, we could color the data frame as a gradient. The color would depend on the cmap parameter where the parameter is accepting colormaps from the matplotlib library.

```python
matrix.style.background_gradient(cmap='icefire')
```

**Visulaization using Seaborn**

**Annotated Heatmap displaying the correlations between the properties**
**bold text**

```python
heatmap_seaborn = sns.heatmap(matrix, annot=True, linewidth = 0.5, linecolor='black', cmap='rocket')
```

**Saving the Heatmap**

```python
plt.savefig('heatmap_seaborn.png', dpi=400)
```
